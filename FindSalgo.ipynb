```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Sky": ["sunny", "sunny", "rainy", "sunny"],
    "AirTemp": ["warm", "warm", "cold", "warm"],
    "Humidity": ["normal", "high", "high", "high"],
    "Wind": ["strong", "strong", "strong", "strong"],
    "Water": ["warm", "warm", "warm", "cool"],
    "Forecast": ["same", "same", "change", "change"],
    "EnjoySport": ["yes", "yes", "no", "yes"]
}

df = pd.DataFrame(data)
print("Training Data:\n")
print(df)
```

```python
X = df.iloc[:, :-1].values   # All attributes
y = df.iloc[:, -1].values   # Target (EnjoySport)

print("\nFeatures:\n", X)
print("\nTarget:\n", y)
```

```python
hypothesis = None

for i in range(len(y)):
    if y[i] == "yes":
        hypothesis = X[i].copy()
        print("\nInitial Hypothesis (from first positive example):")
        print(hypothesis)
        break
```

```python
for i in range(len(y)):
    if y[i] == "yes":
        print(f"\nProcessing positive example {i+1}: {X[i]}")
        for j in range(len(hypothesis)):
            if hypothesis[j] != X[i][j]:
                hypothesis[j] = "?"
        print("Updated Hypothesis:", hypothesis)
```

```python
print("\nFinal Most Specific Hypothesis:")
print(hypothesis)
```
